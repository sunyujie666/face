```python
http_url="https://api-cn.faceplusplus.com/facepp/v3/detect"

key="VPCVp0TurMQ1qehFcqcoH-N4J_ngpZaj"
secret="G2BpBymeAWFV13W5d_J0e9jQyXAzG9xa"

filepath1="F:\\Jupyter\\face\\Untitled Folder\\demo01_pics\\test01.jpg"
data={"api_key":key,"api_secret":secret,"return_gesture":"1"}

files={"image_file":open(filepath1,"rb")}
cap=cv2.VideoCapture(0)

def showVideo():
    while True:
        ret,frame=cap.read()
        cv2.imshow("capture",frame)


def detectFaces():
    while True:
        ret,frame=cap.read()
        #print(frame)

        cv2.imshow("capture",frame)
        cv2.imwrite("F:\\Jupyter\\face\\Untitled Folder\\demo01_pics\\test01.jpg",frame)
        files={"image_file":open(filepath1,"rb")}

        starttime = datetime.datetime.now()
        response = requests.post(http_url,data=data,files=files)
        endtime = datetime.datetime.now()
        print((endtime-starttime).seconds)

        req_con=response.content.decode("utf-8")
        req_dict=JSONDecoder().decode(req_con)
        print(req_dict)

        image=cv2.imread("F:\\Jupyter\\face\\Untitled Folder\\demo01_pics\\test01.jpg")
        face_amount=len(req_dict['faces'])
        if face_amount>0:
            for i in range(face_amount):

                top=req_dict['faces'][i]['face_rectangle']['top']
                left=req_dict['faces'][i]['face_rectangle']['left']
                width=req_dict['faces'][i]['face_rectangle']['width']
                height=req_dict['faces'][i]['face_rectangle']['height']
                cv2.rectangle(image,(left,top),(left+width,top+height),(0,255,0),2)

            cv2.imshow("capture",image)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
        
# 创建两个线程
try:
    _thread.start_new_thread( showVideo,())
    _thread.start_new_thread( detectFaces,())
except:
    print ("Error: 无法启动线程")
```

```python
import _thread
import time

# 为线程定义一个函数
def print_time( threadName, delay):
    count = 0
    while count < 5:
        time.sleep(delay)
        count += 1
        print ("%s: %s" % ( threadName, time.ctime(time.time()) ))

# 创建两个线程
try:
    _thread.start_new_thread( print_time, ("Thread-1", 2, ) )
    _thread.start_new_thread( print_time, ("Thread-2", 4, ) )
except:
    print ("Error: 无法启动线程")

#while 1:
#    pass
```
